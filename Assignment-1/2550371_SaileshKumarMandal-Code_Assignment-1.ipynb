```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('HDI_SouthAsia.csv', low_memory=False)
df['year'] = pd.to_numeric(df['year'], errors='coerce').astype('Int64')
latest_year = int(df['year'].dropna().max())
df = df[df['year']==latest_year].copy()

for col in ['hdi','gross_inc_percap']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df = df.dropna(subset=['country','hdi','gross_inc_percap'])

def iqr_bounds(s):
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    i = q3 - q1
    return q1 - 1.5*i, q3 + 1.5*i

lo_h, hi_h = iqr_bounds(df['hdi'])
lo_g, hi_g = iqr_bounds(df['gross_inc_percap'])
df['outlier_any'] = (df['hdi'].lt(lo_h) | df['hdi'].gt(hi_h) |
                     df['gross_inc_percap'].lt(lo_g) | df['gross_inc_percap'].gt(hi_g))

plt.figure(figsize=(8,5))
sns.scatterplot(data=df[~df['outlier_any']], x='gross_inc_percap', y='hdi',
                color='#4C72B0', alpha=0.7, s=60, label='Non-outlier')
sns.scatterplot(data=df[df['outlier_any']], x='gross_inc_percap', y='hdi',
                color='#C44E52', alpha=0.9, s=80, label='Outlier')
plt.title(f'HDI vs GNI per Capita â€” South Asia (2022) [1.5Ã—IQR rule]')
plt.xlabel('GNI per Capita (PPP, dataset units)'); plt.ylabel('HDI')
plt.grid(True, alpha=0.3); plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Detect outliers for both regions

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset (from Problem 1B)
df = pd.read_csv("HDI_problem1B.csv")

# Region lists
south_asia = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

middle_east = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

# Assign regions
df["Region"] = df["country"].apply(
    lambda x: "South Asia" if x in south_asia
    else ("Middle East" if x in middle_east else None)
)

df = df[df["Region"].notnull()]

def iqr_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

# Detect outliers
df["HDI_outlier"] = iqr_outliers(df["hdi"])
df["GNI_outlier"] = iqr_outliers(df["gross_inc_percap"])

# Mark combined outliers
df["Outlier"] = df["HDI_outlier"] | df["GNI_outlier"]

# Split regions
sa = df[df["Region"] == "South Asia"]
me = df[df["Region"] == "Middle East"]


```

```python

#South Asia :GNI vs HDI with outliers
plt.figure(figsize=(12,5))

# -------- South Asia --------
plt.subplot(1, 2, 1)

plt.scatter(
    sa[~sa["Outlier"]]["gross_inc_percap"],
    sa[~sa["Outlier"]]["hdi"],
    label="South Asia"
)

plt.scatter(
    sa[sa["Outlier"]]["gross_inc_percap"],
    sa[sa["Outlier"]]["hdi"],
    label="Outliers"
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("South Asia: GNI vs HDI with Outliers")
plt.legend()
plt.grid(alpha=0.3)

#Middle East:GNI vs HDI with outliers
plt.subplot(1, 2, 2)

plt.scatter(
    me[~me["Outlier"]]["gross_inc_percap"],
    me[~me["Outlier"]]["hdi"],
    label="Middle East"
)

plt.scatter(
    me[me["Outlier"]]["gross_inc_percap"],
    me[me["Outlier"]]["hdi"],
    label="Outliers"
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Middle East: GNI vs HDI with Outliers")
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
